<h3><b>Scope of a variable</b></h3>
It's important to understand the scope of any variable first before diving into the Object Oriented Programming.<br>
Scope of a variable is defined as the extent of the program code within which the variable can we accessed or declared or worked with. It defines the boundry for any variable within which the variable can be usedf. If we try to access any variable out of it's scope, we will get an error.


<hr><br>Based on the scope of any variable there can be 4 types of variable: Local, Enclosing, Global, Built-in variable.

- **Local Variable:** It is a variable declared inside a particular block of code or any function. The local function can only be used inside the function or the block of code it's defined in and can't be used outside that.

- **Enclosing Variable:** These varaiables are also enclosed by some piece of block or any function, but these are not only restricted to just that block, these variables are also accessible inside other functions or block of code contained by the function/block the variable is defined in. eg: in case of nested loop.   

- **Global Variable:** These are the variables that can be accessed anywhere in the programe.

- **Built-in Variable:** These are the variables that are already defined in the python and have some value. We don't need to define these variables and can be accessed anywhere in the programme.

>**NOTE:** Prefrence order for any variable follows:
*Local Variable > Enclosing Variable > Global Variable > Built-in Variable*. Wen will see all the variables one by one, and the concept of prefrence will become clear when we proceed with the notebook.

<b><h2>Class</h2></b>

Class is like a blueprint for any object, in programming a class is generally used to specify some architecture for similar objects, and then the class is used to create multiple similar or distinct objects with similar properties. 
For example: A class may be represented as structure of a car, while the object can be two different cars.

```python
# To write a class, we start by writing the keyword class and then followed by the class name

class car():

  # Defining the __init__ function for the class (or we may call constructor of the class)

  # We always have to pass the first argument in the constructor as self
  # self reffers to the pointer that points to the instance of the class being created

  def __init__(self, color_of_car, model_of_car):
    self.color = color_of_car   # Sets the property color of the instance of the class car
    self.model = model_of_car   # Sets the property color of the instance of the class car
    self.no_of_wheels = 4       # Sets the property number of wheels as 4 for all the instances
```

**Creating the instance of the class**

```python
# To create an instanc/object of any class, we write the class name and pass the arguments for the init function
car1 = car("white", "Creta")
car2 = car("Black", "Harrier")
```

car1 and car2 are two instances/objects of the class car

**Accessing Variables inside the class**

```python
# The properties/variables of the instances can be accessed by writting the 
# objectName.(variableName)

print(car1.color)
print(car2.model)
```

### **Class Variables and Instance Variables**

For any class, there are two types of variables, the class variables and the instance variables.

 - The instance variables are specific to any instance/object of the class while the class variables are shared between all the instances of a class.
 - The instance variables are defined inside the constructor of the class while the class variables are defined outside the constructor of the class.

```python
class car():
  number_of_wheels = 4 #It's the class variable which is shared by all the instances of the class car

  def __init__(self, color_of_car, model_of_car):
    self.color = color_of_car       # it is an instance variable as it is defined seperatly for each instance   
    self.model = model_of_car       # it is an instance variable as it is defined seperatly for each instance
```

```python
car1 = car("balck", "harrier")
car2 = car("white", "creta")
```

```python
# The instance variable is accessed using the instance name

car1.color
```

```python
# The class variables can be accessed using the class name as well as the instance name

print("Using class name:", car.number_of_wheels)
print("Using instance name:", car1.number_of_wheels)
```

> **NOTE**: Inastance variables should be referred using the class name only and not using the instance name. As the instance variable is shared among the classes, making the change in the instance variable using the class name will change the value for all the instance.

```python
car.number_of_wheels = 5

print(car1.number_of_wheels)
print(car2.number_of_wheels)
```

>**NOTE:** The instance variables can also be defined using the instance name. If we write objectName.variableName = "some value", if no instance variable already exists with that name, a new will be created. **But it's not a good practice and we should avoid this**.

##**Methods in the class**

Just like other vriables, we can create fucntions inside of the class. And these fnctions are known as the *Methods of any class*.
>Note: For any method, we are defining inside the class, we always have to pass first parameter self inside the method.

```python
class car():
  number_of_wheels = 4 

  def __init__(self, color_of_car, model_of_car):
    self.color = color_of_car   
    self.model = model_of_car

  def run(self):    # first argument is always self
    print("Starting the engine!!")
 
  def stop(self):   # first argument is always self
    print("Applying Breaks!!")

  def properties(self):
    print(self.color)   #accessing the instance variable inside the class
    print(self.model)   #accessing the instance variable inside the class
```

```python
car1 = car("white", "creta")

car1.run()  
car1.stop()  
```

##**Scope of a variable in the class**

We can define the scope of any variable in the class by declaring the variables as either *public* or *private* in python.

Public variables can be accessed from anywhere (either inside the class or outside the class), but the private variables can be accessed from inside of the class only.


>**Note:** By default all variables are public, to make them private, we initializae the name of the variables with "__" (double underscores).

```python
class car():
  number_of_wheels = 4 

  def __init__(self, color_of_car, model_of_car):
    self.color = color_of_car       #Public Variable 
    self.__model = model_of_car     #Private Variable

  def run(self):   
    print("Starting the engine!!")
 
  def stop(self):   
    print("Applying Breaks!!")
```

```python
car1 = car("black", "harrier")

print(car1.color)  # accessing the public variable
print(car1.model)    
print(car1.__model)
```

We see if we try to access the private variable from outside the class (using the instance, we are unable to access it)

###**Ways to access the private variables**

```python
# We can access the private variable using the public method from the class
# generally called 'getter' method

class car():
  number_of_wheels = 4 

  def __init__(self, color_of_car, model_of_car):
    self.color = color_of_car       #Public Variable 
    self.__model = model_of_car     #Private Variable

  def run(self):   
    print("Starting the engine!!")
 
  def stop(self):   
    print("Applying Breaks!!")

  def get_model(self):
    return self.__model
```

```python
car1 = car("white", "creta")
print(car1.get_model())
```
